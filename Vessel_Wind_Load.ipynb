# Vessel Wind Load Calculation

This notebook standardizes vessel wind load calculations for marina design, using Tobiasson & Kollmeyer, 2000, as the standard.

The calculations are performed in 'vessel_wind_load.py'
The function checks are performed in 'test_vessel_wind_load.py

**NOTE: Edit only the 'Inputs' section of this notebook.

```python
import vessel_wind_load
import pandas as pd
```

## Inputs

Refer to Figures 15-3 and 15-4 in Tobiasson to determine direction and shielding factors.

Direction factor should be between 1.0 and 3.5
Shielding factor should be between 0.3 and 1.0


<img src="images/fig15_3_direction_factor.png" width="500">

<img src="images/fig15_4_shielding_factor.png" width="500">



```python
## Begin  Editing Here =============================================================

# Vessel Length (ft)
vessel_length_ft = 50 

#Design Wind Speed - Sustained (mph)
design_wind_speed_mph = 45

#Direction Factor
direction_factor = 3.5

#Shielding Factor
shielding_factor = 1.0

## Stop Editing Here ================================================================
```

## Results

```python
# Calculations and formatting here.
beam_ft = vessel_wind_load.calculate_vessel_beam(vessel_length_ft)
height_ft = vessel_wind_load.calculate_vessel_height(vessel_length_ft)
draft_ft = vessel_wind_load.calculate_vessel_draft(vessel_length_ft)
end_area_ft2 = vessel_wind_load.calculate_end_area(vessel_length_ft)
wind_force_kips = vessel_wind_load.calculate_wind_force(vessel_length_ft, direction_factor, shielding_factor, design_wind_speed_mph)


print(f"Calculated Wind Force: {wind_force_kips:.3f} kips")

results = pd.DataFrame({
    "Parameter": [
        "Vessel Length",
        "Estimated Beam",
        "Estimated Height Above Water",
        "Estimated Draft",
        "Estimated End Area",
        "Design Wind Speed",
        "Direction Factor",
        "Shielding Factor",
        "Wind Force",
    ],
    "Value": [
        round(vessel_length_ft, 0),
        round(beam_ft, 1),
        round(height_ft,1),
        round(draft_ft, 1),
        round(end_area_ft2, 1),
        round(design_wind_speed_mph, 0),
        round(direction_factor, 1),
        round(shielding_factor, 1),
        round(wind_force_kips, 3),
    ],
    "Units": [
        "ft",
        "ft",
        "ft",
        "ft",
        "ft²",
        "mph",
        "-",
        "-",
        "kips",
    ]
})

results
```
